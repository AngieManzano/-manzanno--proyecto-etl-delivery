3. Estrategia de Menú por Ciudad: Para las 3 ciudades más populares obtenidas
en la pregunta 1, calcule el precio promedio de los platos en el menú. ¿En
qué ciudad es más caro comer en promedio?

```python
query_3 = """
SELECT
    ciudad,
    ROUND(AVG(price), 2) AS precio_promedio
FROM master_food_data
WHERE ciudad IN ('Milwaukee', 'Seattle', 'Lynnwood')
GROUP BY ciudad
ORDER BY precio_promedio DESC;
"""
```

```python
with motor_sql.connect() as conn:
    promedio_total_restaurantes_1 = pd.read_sql(text(query_3), conn)
    print(promedio_total_restaurantes_1)
```

```python
sns.set_theme(style="whitegrid")
plt.figure()


ax = sns.barplot(
    x='ciudad',
    y='precio_promedio',
    data=promedio_total_restaurantes_1,
    palette='Oranges_r',
    hue='ciudad',
    legend=False
)



plt.title('Precio Promedio del Menú en las 3 Ciudades Principales', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Ciudad', fontsize=12)
plt.ylabel('Precio Promedio por Plato', fontsize=12)
plt.ylim(0, promedio_total_restaurantes_1['precio_promedio'].max() * 1.15)

plt.tight_layout()
plt.show()
```

4. La Pregunta del Millón (Correlación Precio-Calidad): ¿Un restaurante más
caro garantiza una mejor experiencia para el usuario? Escriba un query que
calcule el puntaje (score) promedio agrupado por rango_de_precios. Utilice un
gráfico de líneas o de barras para demostrar visualmente al CEO si la calidad
sube a medida que aumenta la categoría de precio.

```python
query_4 = """

SELECT
    price_range,
    ROUND(AVG(score), 2) AS score_promedio,
    COUNT(DISTINCT id_restaurantes) AS total_restaurantes
FROM  master_food_data
WHERE price_range IS NOT NULL AND score IS NOT NULL
GROUP BY  price_range
;
"""
```

```python
with motor_sql.connect() as conn:
    promedio_restaurante_puntaje= pd.read_sql(text(query_4), conn)
    print(promedio_restaurante_puntaje)
```

```python
ax = sns.barplot(
    x='price_range',
    y='score_promedio',
    data=promedio_restaurante_puntaje,
    palette='Purples'
)

for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f} ★",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 8),
        textcoords='offset points',
        fontweight='bold',
        fontsize=11
    )

plt.title('Score Promedio por Rango de Precios', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Categoría de Precio', fontsize=12)
plt.ylabel('Puntaje Promedio', fontsize=12)


plt.ylim(0, 5.0)

plt.tight_layout()
plt.show()
```

Se encontró que no necesariamente si el restauranante es más caro es mejor la experiencia para el ususario. Ya que la categoria de "Economico" supero en experiencia del usuario a la categoria de "Caro". La categoria de "Economico" y "Moderadamente caro" obtuvieron el mayor puntaje, haciendo que ambos sean los restaurantes con mejor experiencia de ususario.
